```python
import pandas as pd
```

```python
df = pd.read_csv(r'data/rawdata/1_A_btc_price_data_2013-05-01_2019-11-01.csv', parse_dates=['Timestamp'])
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
# converting to date_time format
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df.head()
```

```python
# Set the 'Timestamp' column as the index
df.set_index('Timestamp', inplace=True)

# Resample the data to hourly frequency
df_hourly = df.resample('H').mean()
```

```python
# To keep the Timestamp as a column instead of an index, we can reset the index
df_hourly.reset_index(inplace=True)
df_hourly.head(10)
```

```python
df_hourly.shape
```

```python
df_hourly.columns
```

```python
# Export the hourly data to a CSV file
df_hourly.to_csv('data/btc_price_hourly_data_2013-05-01_2019-11-01.csv', index=False)
```
